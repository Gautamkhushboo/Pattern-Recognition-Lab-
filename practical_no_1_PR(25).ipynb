```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances, manhattan_distances
```

```python
customers = {
    "C1": [5000, 5, 4, 30],
    "C2": [5200, 6, 5, 35],
    "C3": [2000, 2, 3, 10]
}

data = np.array(list(customers.values()))
names = list(customers.keys())

print("Customer Data:\n", data)
```

```python
euclidean = euclidean_distances(data)

print("Euclidean Distance:\n")
print(euclidean)
```

```python
from sklearn.metrics import pairwise_distances

manhattan = pairwise_distances(data, metric='manhattan')

print("Manhattan Distance:\n")
print(manhattan)
```

```python
cosine = cosine_similarity(data)

print("Cosine Similarity:\n")
print(cosine)
```

```python
def most_similar(sim_matrix, names):
    n = len(names)
    for i in range(n):
        best_score = -1
        best_match = None

        for j in range(n):
            if i != j:
                if sim_matrix[i][j] > best_score:
                    best_score = sim_matrix[i][j]
                    best_match = names[j]

        print(f"{names[i]} → most similar to {best_match} (score: {best_score:.2f})")
```

```python
most_similar(cosine, names)
```
